```python
pos = np.ones(100) * 0.01
dt = 1e-3
pid = PID(0.001, 0, 0, 0)

for i in range(len(pos)-1):
    pos[i+1] = evolve_system(pos[i], 0.01, 1)
    pos[i+1] = pid(np.sin(pos[i+1])**2, dt)

plt.plot(np.sin(pos)**2)
```
